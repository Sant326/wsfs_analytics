## Observaciones – Tabla Customers

- La tabla contiene 80 registros y 10 columnas.
- No se detectaron duplicados exactos en los registros.
- Se identificaron valores nulos en campos no críticos:
  - phone (5 registros)

- Las claves `customer_id` y `customer_number` no presentan valores nulos.

Se detecta repetición de valores en `customer_name`, lo que indica que este campo no puede utilizarse como identificador único.  
Para relaciones entre tablas se recomienda utilizar `customer_number`.


```python
df_customers.groupby("customer_name")["customer_number"].nunique().sort_values(ascending=False).head(10)
```

### Análisis

- Vemos que los datos tienen distinto ID, distinto customer_number, distintas fechas de creación, distintos departamentos ciudades y teléfonos, pero contacto por correo iguales. En un contexto real, este patrón podría significar que son sucursales de una misma firma, pero al tratarse de datos creados con IA y no estar seguros, tomé la decisión de tratarlos por customer_number como clientes independientes.
